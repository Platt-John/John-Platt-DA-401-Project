```python
#Module to load in data with
import pandas as pd
#Module to calculate distance
from geopy import distance
```

```python
def starbucks_distance(N,data):
  '''
  Given user inputs on location, find the N nearest starbucks.

    Parameters:
    N - int - number of nearest starbucks to return
    data - object - table with starbucks coordinates

  Return:
  Return the N nearest starbucks
  '''
  #Remove missing values from data's longitude and latitude columns
  data.dropna(subset=['Latitude','Longitude'],inplace=True)
  #User inputs latitude/longitude
  user_latitude=float(input("What is your latitude? "))
  user_longitude=float(input("What is your longitude? "))
  #Create user coordinate
  user_point=(user_latitude,user_longitude)
  #Initialize distance list
  distance_list=[]
  #Iterate through latitudes and longitudes of the Starbucks data
  for latitude, longitude in zip(data['Latitude'],data['Longitude']):
  #Create Starbucks coordinate
    point=(latitude,longitude)
  #Append distance between user coordinate and starbucks coordinate to distance list
    distance_list.append(distance.distance(user_point,point).km)
  #Create new distance column in data
  data['distance_km']=distance_list
  #Return closest starbucks
  return data.sort_values(by='distance_km').head(N)
```

```python
#Make sure csv is in same folder as this .ipynb file
starbucks_data=pd.read_csv('starbucks_locations.csv')
```

```python
#Return information about the 5 nearest starbucks
starbucks_distance(5,starbucks_data)
```
